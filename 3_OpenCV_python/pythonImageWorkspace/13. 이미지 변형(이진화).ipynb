## Trackbar
값 변화에 따른 변형 확인

```python
import cv2
path = "C:\\Users\\User\\Desktop\\code\\gitFolder\\myStudy\\3_OpenCV_python\\pythonImageWorkspace\\Data"

def empty(pos):
    # print(pos)
    pass

img = cv2.imread(path + '/book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow(name)

# bar 이름, 창의 이름, 초기값, 최대값, 이벤트 처리
cv2.createTrackbar('threshold', name, 127, 255, empty)

while True:
    # bar 이름, 창의 이름
    thresh = cv2.getTrackbarPos('threshold', name)
    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)
    
    if not ret:
        break

    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```
